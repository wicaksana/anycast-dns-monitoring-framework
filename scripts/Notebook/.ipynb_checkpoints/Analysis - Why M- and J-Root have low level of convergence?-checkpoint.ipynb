# M-Root

This is what M-Root looks like

![alt text](../analysis/figs/png/convergence_over_time_m.png)

The hypothesis is that M-Root has different routing policies for IPv4 and IPv6, especially heavy-reliance to HE for IPv6 traffic.

**Analysis idea:** For each date, find out the upstream providers for both IPv4 and IPv6

```python
%matplotlib inline
from ggplot import *
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
import csv
from pandas import DataFrame
from pymongo import MongoClient, ASCENDING, DESCENDING
from bson.code import Code
from collections import Counter

import matplotlib.dates as mdates
```

```python
client = MongoClient()
anycast = client.anycast_monitoring

root_list = 'acdfijklm'

# reducer collects penultimate ASes
reducer = Code("""
    function(obj, prev) {
        prev.pu_v4.push(obj.path4[obj.path4.length - 2]);
        prev.pu_v6.push(obj.path6[obj.path6.length - 2]);
    }
"""
)

def dateparse (time_in_secs):    
    return datetime.fromtimestamp(float(time_in_secs))

container2 = dict()

for root in root_list:
    coll = anycast['{}_root'.format(root)]
    res = coll.group(key={'timestamp':1}, condition={}, initial={'pu_v4': [], 'pu_v6': []}, reduce=reducer)
    container2[root] = DataFrame(res)
    
    container2[root]['timestamp'] = pd.to_datetime(container2[root]['timestamp'] * 1000000000)
    container2[root] = container2[root].set_index(['timestamp'])
    
#     container2[root]['coll_pu_v4'] = pd.Series()
#     container2[root]['coll_pu_v6'] = pd.Series()
    
#     for item in container2[root].iterrows():
#         item[1]['coll_pu_v4'] = Counter(item[1]['pu_v4']).most_common()[:3]
#         item[1]['coll_pu_v6'] = Counter(item[1]['pu_v6']).most_common()[:3]
```

```python
container2['m']
```

```python
from collections import defaultdict

mc_4 = list()
mc_6 = list()

# fig, ax = plt.subplots()

for item in container2['m'].iterrows():
    v4 = Counter(item[1]['pu_v4']).most_common()[:1]
    v6 = Counter(item[1]['pu_v6']).most_common()[:1]
    
    # find out list of the most common PU of IPv4
    for item in v4:
          mc_4.append(item[0])
    for item in v6:
          mc_6.append(item[0])

print(Counter(mc_4).most_common())
print(Counter(mc_6).most_common())
    
```

Show most dominant upstream providers over the time

```python
for item in container2['m'].iterrows():
    v4 = Counter(item[1]['pu_v4']).most_common()[:3]
    v6 = Counter(item[1]['pu_v6']).most_common()[:3]
    
    print('{}: {}\t\t\t{}'.format(item[0], v4, v6))
```

Screw it. Let us just plot the top IPv4 and IPv6 upstream providers used by M-Root

```python
container2['m']['3257'] = pd.Series()
container2['m']['6939'] = pd.Series()
container2['m']['24785'] = pd.Series()

for index, item in container2['m'].iterrows():
    counter4 = Counter(item['pu_v4'])
    counter6 = Counter(item['pu_v6'])
    total = len(item['pu_v4'])
    
    container2['m'].loc[index, '3257'] = counter4[3257.0] / total
    container2['m'].loc[index, '6939'] = counter6[6939.0] / total
    container2['m'].loc[index, '24785'] = counter4[24785.0] / total
    
#     print('{} {} {}'.format(counter4[3257.0], counter6[6939.0], total))
    
container2['m']
```

plot it

```python
import matplotlib.dates as mdates

# date formatter
years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
yearsFmt = mdates.DateFormatter('%Y')

# http://matplotlib.org/examples/api/date_demo.html
fig, ax = plt.subplots()
# container2['m'].plot(ylim=(0,1), figsize=(12,4), ax=ax)
container2['m']['3257'].plot(ylim=(0,1), figsize=(12,4), ax=ax, label='3257 (IPv4)')
container2['m']['6939'].plot(ylim=(0,1), figsize=(12,4), ax=ax, label='6939 (IPv6)', linestyle='--')
container2['m']['24785'].plot(ylim=(0,1), figsize=(12,4), ax=ax, label='24785 (IPv4)')

# format the ticks
ax.xaxis.set_major_locator(years)
ax.xaxis.set_major_formatter(yearsFmt)
ax.xaxis.set_minor_locator(months)

# datemin = datetime.date(r.date.min().year, 1, 1)
# datemax = datetime.date(r.date.max().year + 1, 1, 1)
# ax.set_xlim(datemin, datemax)

ax.format_xdata = mdates.DateFormatter('%Y-%m-%d')

ax.grid(True)
plt.xlabel('')
plt.legend(loc='upper left')

plt.show()
```
