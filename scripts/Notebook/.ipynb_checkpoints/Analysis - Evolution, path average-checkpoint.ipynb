```python
from pymongo import MongoClient, ASCENDING, DESCENDING
from bson.code import Code

%matplotlib inline
from ggplot import *
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime
import csv
from pandas import DataFrame
```

# Some tests

```python
a_root = anycast.a_root
c_root = anycast.c_root
```

```python
for item in a_root.find().sort('timestamp', ASCENDING):
    print(item['timestamp'])
```

```python
reducer = Code("""
    function(obj, prev) {
        if(obj.path4.toString() != obj.path6.toString()) {
            prev.different++;
        } else {
            prev.identical++;
        }
    }
"""
)

results = c_root.group(key={'timestamp':1}, condition={}, initial={'identical': 0, 'different': 0}, reduce=reducer)

for result in results:
    print(result)
```

```python
plot_data = DataFrame(results)
plot_data = plot_data.set_index(['timestamp'])

plot_data['total'] = pd.Series()
plot_data['diff_pct'] = pd.Series()
plot_data['ident_pct'] = pd.Series()

for item in plot_data.iterrows():
    item[1]['total'] = item[1]['different'] + item[1]['identical']
    item[1]['diff_pct'] = item[1]['different'] / item[1]['total'] * 100
    item[1]['ident_pct'] = item[1]['identical'] / item[1]['total'] * 100

# plot_data
fig, ax = plt.subplots()

plot = plot_data[['diff_pct', 'ident_pct']].plot.bar(stacked=True, figsize=(14, 5), width=1.0, ax=ax, alpha=0.6)
ax.legend(['different', 'identical'])

n = 6

ticks = plot.xaxis.get_ticklocs()
ticklabels = [datetime.fromtimestamp(float(l.get_text())).strftime('%Y-%m-%d') for l in plot.xaxis.get_ticklabels()]
plot.xaxis.set_ticks(ticks[::n])

plt.title('Composition of peers for C-Root Server')
plt.ylabel('Percentage (%)')

plt.show()
```

# Convergence Level

```python
root_list = 'acdfijklm'

container = dict()

for root in root_list:
    coll = anycast['{}_root'.format(root)]
    res = coll.group(key={'timestamp':1}, condition={}, initial={'identical': 0, 'different': 0}, reduce=reducer)
    
    container[root] = DataFrame(res)
    
    container[root] = container[root].set_index(['timestamp'])

    container[root]['total'] = pd.Series()
    container[root]['diff_pct'] = pd.Series()
    container[root]['ident_pct'] = pd.Series()

    for item in container[root].iterrows():
        item[1]['total'] = item[1]['different'] + item[1]['identical']
        item[1]['diff_pct'] = item[1]['different'] / item[1]['total'] * 100
        item[1]['ident_pct'] = item[1]['identical'] / item[1]['total'] * 100


fig, ax = plt.subplots()

for index, item in enumerate(container):
    plot = container[item]['ident_pct'].plot(ylim=(0,100),figsize=(14,8), ax=ax, label='{}-root'.format(root_list[index]))

plt.legend(loc='upper left')

plt.show()
```

```python
def dateparse (time_in_secs):    
    return datetime.fromtimestamp(float(time_in_secs))

container2 = dict()

for root in root_list:
    coll = anycast['{}_root'.format(root)]
    res = coll.group(key={'timestamp':1}, condition={}, initial={'identical': 0, 'different': 0}, reduce=reducer)
    container2[root] = DataFrame(res)
    
    container2[root]['timestamp'] = pd.to_datetime(container2[root]['timestamp'] * 1000000000)
    container2[root] = container2[root].set_index(['timestamp'])
    
    container2[root]['total'] = pd.Series()
    container2[root]['diff_pct'] = pd.Series()
    container2[root]['ident_pct'] = pd.Series()

    for item in container2[root].iterrows():
        item[1]['total'] = item[1]['different'] + item[1]['identical']
        item[1]['diff_pct'] = item[1]['different'] / item[1]['total'] * 100
        item[1]['ident_pct'] = item[1]['identical'] / item[1]['total']


fig, ax = plt.subplots(nrows=3, ncols=3, sharey=True)
fig.text(0, 0.5, 'Fraction of peers with identical IPv4/IPv6 paths', va='center', rotation='vertical')
fig.text(1, 0.5, 'Amount of peers', va='center', rotation='vertical')
# fig.text(0.5, 1, 'Percentage of Peers with Identical IPv4/IPv6 Path', horizontalalignment='center', fontsize=18)

counter = 0
for i in range(0, 3):
    for j in range(0, 3):
        ax2 = ax[i,j].twinx()
        container2[root_list[counter]]['ident_pct'].plot(ylim=(0,1), xlim=(pd.Timestamp('2008-02-1'), pd.Timestamp('2016-06-01')), ax=ax[i,j], figsize=(12, 8), label="fraction")
        container2[root_list[counter]]['total'].plot(color='r', alpha=0.35, xlim=(pd.Timestamp('2008-02-1'), pd.Timestamp('2016-06-01')), ax=ax2, figsize=(12, 8), label=r"peers")
        ax[i,j].set_title('{}-Root'.format(root_list[counter].upper()))
        if root_list[counter] == 'l':
            ax[i,j].legend(loc='lower left')
            ax2.legend(loc='lower right')
        ax[i,j].grid(True)
        ax[i,j].set_xlabel('')
#         ax[i,j].text(pd.Timestamp('2008-05-1'), 90, '{}-Root'.format(root_list[counter].upper()), bbox={'facecolor':'white', 'pad':5})
        if i != 2:
            ax[i,j].xaxis.set_ticklabels([])
        counter += 1

plt.tight_layout()
plt.show()
```

```python
reducer = Code("""
    function(obj, prev) {
        prev.len4.push(obj.path4.length);
        prev.len6.push(obj.path6.length);
    }
"""
)

root = 'a'
container = dict()
coll = anycast['a_root']

res = coll.group(key={'timestamp':1}, condition={'$where': 'this.path4.toString() != this.path6.length.toString()'}, initial={'len4': [], 'len6': []}, reduce=reducer)
container[root] = DataFrame(res)
container[root]['timestamp'] = pd.to_datetime(container[root]['timestamp'] * 1000000000)
container[root] = container[root].set_index(['timestamp'])

fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)

ipv4 = DataFrame()
ipv6 = DataFrame()

for item in container[root].iterrows():
    s = DataFrame({item[0]: item[1]['len4']})
    t = DataFrame({item[0]: item[1]['len6']})
    ipv4 = pd.concat([ipv4, s], axis=1)
    ipv6 = pd.concat([ipv6, t], axis=1)

plot_v4 = ipv4.plot.box(ax=axes[0], figsize=(10,5), ylim=(1.5, 9.5))
plot_v6 = ipv6.plot.box(ax=axes[1], figsize=(10,5), ylim=(1.5, 9.5))

n = 6

axes[0].text(3, 8, 'IPv4', fontsize=20, bbox={'facecolor':'white', 'pad':5})
axes[0].grid(True)

ticks = axes[1].xaxis.get_ticklocs()
ticklabels = [l.get_text().split(' ')[0] for l in axes[1].xaxis.get_ticklabels()]
axes[1].xaxis.set_ticks(ticks[::n])
axes[1].text(3, 8, 'IPv6', fontsize=20, bbox={'facecolor':'white', 'pad':5})
axes[1].xaxis.set_ticklabels(ticklabels[::n], rotation=15)

axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## All peers

```python
reducer = Code("""
    function(obj, prev) {
        prev.len4.push(obj.path4.length);
        prev.len6.push(obj.path6.length);
    }
"""
)

def format_date(dt):
    try:
        dt = dt.split(' ')[0]
        year, month, day = dt.split('-')
        returned = day + '/' + month + '/' + year[2:]
    except ValueError:
        returned = dt
        print('error: {}'.format(dt))
    return returned


for root in root_list:
    print('{}-Root:'.format(root))
    
    container = dict()
    coll = anycast['{}_root'.format(root)]

#     res = coll.group(key={'timestamp':1}, condition={'$where': 'this.path4.toString() != this.path6.length.toString()'}, initial={'len4': [], 'len6': []}, reduce=reducer)
    res = coll.group(key={'timestamp':1}, condition={}, initial={'len4': [], 'len6': []}, reduce=reducer)
    container[root] = DataFrame(res)
    container[root]['timestamp'] = pd.to_datetime(container[root]['timestamp'] * 1000000000)
    container[root] = container[root].set_index(['timestamp'])

    fig, axes = plt.subplots(nrows=2, ncols=1, sharex=True)

    ipv4 = DataFrame()
    ipv6 = DataFrame()

    for item in container[root].iterrows():
        s = DataFrame({item[0]: item[1]['len4']})
        t = DataFrame({item[0]: item[1]['len6']})
        ipv4 = pd.concat([ipv4, s], axis=1)
        ipv6 = pd.concat([ipv6, t], axis=1)
        
    # find median
    diff_ipv4_list = list()
    for item in container[root].iterrows():
        diff_ipv4_list = diff_ipv4_list + item[1]['len4']
    
    diff_ipv4_df = DataFrame(diff_ipv4_list)
    median4 = diff_ipv4_df.median(skipna=True)
    
    diff_ipv6_list = list()
    for item in container[root].iterrows():
        diff_ipv6_list = diff_ipv6_list + item[1]['len6']
    
    diff_ipv6_df = DataFrame(diff_ipv6_list)
    median6 = diff_ipv6_df.median(skipna=True)
    
    # plot IPv4 path length median as a green line and display the value
    plot_v4 = ipv4.plot.box(ax=axes[0], figsize=(14,3), ylim=(1.5, 9.5))
    axes[0].axhline(color='g', linewidth=4, alpha=0.4, y=float(median4))
    axes[0].text(1.01, 
                 median4.get_values()[0] / 10, # normalize 
                 'median: {0:.2f}'.format(median4.get_values()[0]), 
                 fontsize=8, horizontalalignment='center', 
                 verticalalignment='center', 
                 transform= axes[0].transAxes, 
                 rotation=90)
    
    # plot IPv6 path length mean as a green line
    plot_v6 = ipv6.plot.box(ax=axes[1], figsize=(14,3), ylim=(1.5, 9.5))
    axes[1].axhline(color='g', linewidth=4, alpha=0.4, y=float(median6))
    axes[1].text(1.01, 
                 median6.get_values()[0] / 10, # normalize 
                 'median: {0:.2f}'.format(median6.get_values()[0]), 
                 fontsize=8, horizontalalignment='center', 
                 verticalalignment='center', 
                 transform= axes[1].transAxes, 
                 rotation=90)
    
    n = 6

    axes[0].text(0.025, 0.85, 'IPv4', fontsize=12, bbox={'facecolor':'white', 'pad':3}, horizontalalignment='center', verticalalignment='center', transform= axes[0].transAxes)
    axes[0].grid(True)

    ticks = axes[1].xaxis.get_ticklocs()
    ticklabels = [format_date(l.get_text()) for l in axes[1].xaxis.get_ticklabels()]
    axes[1].xaxis.set_ticks(ticks[::n])
    axes[1].text(0.025, 0.85, 'IPv6', fontsize=12, bbox={'facecolor':'white', 'pad':3}, horizontalalignment='center', verticalalignment='center', transform= axes[1].transAxes)
    axes[1].xaxis.set_ticklabels(ticklabels[::n])

    axes[1].grid(True)

    # plt.locator_params(axis='x',nbins=10)
    plt.tight_layout()
    plt.show()
    
    # save figure
    fig.savefig('../analysis/figs/eps/convergence_over_time_{}.eps'.format(root), format='eps', dpi=1000)
    fig.savefig('../analysis/figs/png/convergence_over_time_{}.png'.format(root))
```
